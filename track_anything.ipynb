# Track-Anything

Interactive video object segmentation and tracking powered by [Segment Anything (SAM)](https://github.com/facebookresearch/segment-anything) + optical flow.

**Workflow:**
1. Upload a video via the Gradio UI
2. Click on the object you want to track in the first frame
3. The model segments and tracks it through all subsequent frames
4. Download the annotated output video
